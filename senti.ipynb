```python
import string
from collections import Counter
import matplotlib.pyplot as plt
import GetOldTweets3 as got
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
def get_tweets():
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('black lives matter')\
                                           .setSince("2020-05-1")\
                                           .setUntil("2020-06-1")\
                                           .setMaxTweets(2000)
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    tweet_text=[[tweet.text]for tweet in tweets]
    return tweet_text
```

```python
text=""
tweet_text=get_tweets()
for i in range(len(tweet_text)):
    text=tweet_text[i][0]+"\n"+text
print(text)
```

```python
lowertext=text.lower()
print(lowertext)
```

```python
clean=lowertext.translate(str.maketrans('','',string.punctuation))
print(clean)
```

```python
token=word_tokenize(clean,'english')
print(token)
```

```python
def sentiment(sentiment_cont):
    score=SentimentIntensityAnalyzer().polarity_scores(sentiment_cont)
    negative=score['neg']
    positive=score['pos']
    neutral=score['neu']
    if (negative<positive):
        print(positive)
        print("positive statement")
    elif(positive<negative):
        print(negative)
        print("negative statement")
    else:
        print(neutral)
        print('neutral statement')
    langs = ['positive', 'negative', 'neutral']
    em = [positive,negative,neutral]
    plt.pie(em, labels = langs,autopct='%1.2f%%')
sentiment(clean)
```

```python
st=[]
for wo in token:
    if wo not in stopwords.words('english'):
        st.append(wo)
print(st)
```

```python
emotion_list=[]
with open('emotions.txt','r') as file:
    for line in file:
        clear_line=line.replace('\n','').replace(',','').replace("'",'').strip()
        word,emotion=clear_line.split(':') 
        if word in st:
            emotion_list.append(emotion)
```

```python
co=Counter(emotion_list)
co
```

```python
fig ,ax1=plt.subplots(figsize=(20,10))
ax1.bar(co.keys(),co.values())
fig.autofmt_xdate()
ax1.set_ylabel('Frequency')
ax1.set_xlabel('Emotions used')
plt.subplots_adjust()
plt.show()
```
